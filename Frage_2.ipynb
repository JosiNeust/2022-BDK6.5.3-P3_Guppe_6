# Frage 2: In welchem Jahre hat Tim Burton bei den meisten Filmen Regie geführt?

# Import der Bibliotheken

```python
import urllib.request
import gzip
import pandas as pd
import matplotlib.pyplot as plt
```

# Import und Aufbereitung der Datensätze

```python
# Speichern der benötigten URLs
url_data_crew = "https://datasets.imdbws.com/title.crew.tsv.gz"
url_data_year = "https://datasets.imdbws.com/title.basics.tsv.gz"
```

```python
# Zusammenfassen der benötigten Daten (Name der Datei, URL)
title_crew_file = "title_crew_file.csv", url_data_crew
title_year_file = "title_year_file.csv", url_data_year
```

```python
# Download der Daten auf lokalem Gerät
urllib.request.urlretrieve(title_crew_file[1], title_crew_file[0])
urllib.request.urlretrieve(title_year_file[1], title_year_file[0])
```

```python
# Auslesen der Informationen zu Crew und Speicherung in einer Variable
crew = pd.read_csv(title_crew_file[0], compression = 'gzip', sep = "\t")

# Auslesen der Informationen zu Jahren und Speicherung in einer Variable
year = pd.read_csv(title_year_file[0], compression = 'gzip', sep = "\t")
# Dtype Warnung aufgrund unterschiedlicher Datentypen im Datensatz
```

# Suche nach dem Director Timothy Walter Burton im Datensatz

```python
crew
```

```python
# Wenn der Director identisch mit "nm0000318" (Tim Burton) ist, werden die Daten aus dem Datensatz in der Variable gespeichert
# Es wird in crew geschaut, ob im Label "directors" von crew "nm0000318" vorkommt
# - Trifft dies zu, werden alle Informationen der Zeile in der Variable gespeichert
crew_Tim_Burton = crew[crew["directors"] == "nm0000318"]
```

```python
crew_Tim_Burton
```

## Suche nach den entsprechenden Jahren

```python
year
```

```python
# Verknüpfung der beiden Datensätze mittels merge
crew_Tim_Burton_with_years = crew_Tim_Burton.merge(year, on = "tconst", how = "left")
```

```python
crew_Tim_Burton_with_years
```

### Ergebnis in chronologischer Reihenfolge

```python
# Auflistung der Jahre in absteigender Reihenfolge nach Häufigkeit der Veröffentlichungen von Tim Burton
crew_Tim_Burton_with_years["startYear"].value_counts()
```
